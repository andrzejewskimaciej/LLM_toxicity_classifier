```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('BAN-PL_1.csv')
```

```python
df1.drop(columns='id', inplace=True)
df1.rename(columns={'Text' : 'text', 'Class' : 'bin_toxicity'}, inplace=True)
df1['lang'] = 'pl'
df1
```

```python
df2 = pd.read_parquet('civil_comments_test.parquet')
df2['lang'] = 'eng'
df2['bin_toxicity'] = (df2['toxicity'] > 0.5).astype(int)
df2
```

```python
df1_ = pd.concat([df1.loc[df1.bin_toxicity==0][:100], df1.loc[df1.bin_toxicity==1][:100]], axis=0)
df1_
```

```python
df2_ = pd.concat([df2.loc[df2.bin_toxicity==0][:100], df2.loc[df2.bin_toxicity==1][:100]], axis=0)
df2_
```

```python
df = pd.concat([df1_, df2_], axis=0).reset_index(drop=True)
df
```

```python
df.to_csv("test_data.csv", index=False)
```
