```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('BAN-PL_1.csv')
df1
```

```python
df1.drop(columns='id', inplace=True)
df1.rename(columns={'Text' : 'text', 'Class' : 'bin_toxicity'}, inplace=True)
df1['lang'] = 'pl'
```

```python
df2 = pd.read_parquet('civil_comments_test.parquet')
df2['lang'] = 'eng'
df2
```

```python
np.random.seed(0)
df = pd.concat([df1.iloc[np.random.choice(df1.shape[0], size=200), :], df2.iloc[np.random.choice(df2.shape[0], size=200), :]], axis=0)
```

```python
df = df.reset_index(drop=True)
df
```

```python
df.to_csv("test_data.csv", index=False)
```
